## Data Preprocessing

#### Convert Categorical to Numerical

```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
# Features & Target
X = df[['Age', 'Gender', 'Monthly_Income', 'Spending_Score']]
y = df['Purchased']
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

#### Machine Learning Model (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

#### Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#### Save the Final Model

```python
import pickle

pickle.dump(model, open("customer_model.pkl", "wb"))
```

#### Create Streamlit App (app.py)

```python
import streamlit as st
import pickle
import numpy as np

# Load model
model = pickle.load(open("customer_model.pkl", "rb"))

# Page title
st.set_page_config(page_title="Customer Purchase Prediction")

st.title(" Customer Purchase Prediction")
st.write("Predict whether a customer will purchase based on their details.")

# User inputs
age = st.number_input("Age", min_value=18, max_value=70, value=30)

gender = st.selectbox("Gender", ["Male", "Female"])
gender = 1 if gender == "Female" else 0

income = st.number_input(
    "Monthly Income",
    min_value=10000,
    max_value=100000,
    value=40000
)

score = st.slider("Spending Score", 0, 100, 50)

# Prediction
if st.button("Predict Purchase"):
    input_data = np.array([[age, gender, income, score]])
    prediction = model.predict(input_data)

    if prediction[0] == 1:
        st.success(" Customer is likely to PURCHASE")
    else:
        st.error(" Customer is NOT likely to purchase")
```
